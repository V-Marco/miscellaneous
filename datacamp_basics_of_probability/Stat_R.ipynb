```r
library(ggplot2)

encoding = "utf-8"

ggplot(data = data.frame(x = c(0, 30)), aes(x)) +
  stat_function(fun = dchisq, n = 100, args = list(df = 5), size = 1, aes(color = 'line1')) +
  stat_function(fun = dchisq, n = 100, args = list(df = 2), size = 1, aes(color = 'line2')) + 
  scale_y_continuous(breaks = NULL) +
  ggtitle("Привет") +
  theme(plot.title = element_text(hjust = 0.5)) +
  geom_vline(xintercept = 6, linetype='dotted')
```

```r
library(ggplot2)

encoding = "utf-8"

ggplot(data = data.frame(x = c(-4, 4)), aes(x)) +
  stat_function(fun = dnorm, n = 100, args = list(mean = 0, sd = 1), aes(color='E(X)=0, Var(X)=1')) + ylab("f(x)") +
  scale_y_continuous(breaks = NULL) +
  ggtitle("Плотность нормального распределения") +
  theme(plot.title = element_text(hjust = 0.5))
```
