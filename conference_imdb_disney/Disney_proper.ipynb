```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dis_data = pd.read_excel("Disney_nedeed.xlsx", )
```

```python
dis_data.head()
```

```python
dis_stock = pd.read_csv("DIS.csv")
```

```python
dis_stock.head()
```

```python
dis_stock.columns = ['Date_mb', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
```

```python
mb = dis_data['Date_mb']
a = pd.to_datetime(mb)
dis_data['Date_mb'] = a
```

```python
twa = dis_data['Date_twa']
b = pd.to_datetime(twa)
dis_data['Date_twa'] = b
```

```python
dis_stock['Date_mb'] = pd.to_datetime(dis_stock['Date_mb'])
```

```python
dis_stock.info()
```

```python
s1 = pd.merge(dis_data, dis_stock, how='inner', on=['Date_mb'])
```

```python
s1
```

```python
s1.to_excel('disney_mb.xlsx')
```

```python
dis_new_info = pd.read_excel('disney_mb.xlsx')
```

```python
dis_new_info['Date_twa'] = pd.to_datetime(dis_new_info['Date_twa'])
```

```python
dis_new_info.info()
```

```python
dis_stock.head()
```
