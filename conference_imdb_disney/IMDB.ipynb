```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(normalize=True)
```

```python
Y = imdb['Rating']
```

```python
imdb.columns
```

```python
X = imdb.loc[:,'Animation':'Family']
X.shape
```

```python
regressor.fit(X, Y)
```

```python
def gen_prediction(first, second, third):
    a = list(np.zeros(20))
    b = []
    for el in a:
        b.append(int(el))
    for element in add_columns:
        if element == first:
            b[add_columns.index(first)] = 1
        if element == second:
            b[add_columns.index(second)] = 1
        if element == third:
            b[add_columns.index(third)] = 1
    return np.array(b)
```

```python
a = gen_prediction('Action', 'Adventure', 'Fantasy').reshape(1,-1)
y_predict = regressor.predict(a)
y_predict
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(regressor, X, Y, scoring='neg_mean_squared_error', cv=100).mean()
```

```python
regres_rev = LinearRegression(normalize=True)
```

```python
Y_r = imdb['Revenue (Millions)'].fillna(imdb['Revenue (Millions)'].mean())
```
